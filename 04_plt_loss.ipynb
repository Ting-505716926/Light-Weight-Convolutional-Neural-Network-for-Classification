```python
import os
import pandas as pd
from matplotlib import pyplot as plt
```

```python
def loss_visualize(epoch, CL_train_loss, CL_val_loss, ML_train_loss, ML_val_loss, CS_train_loss, CS_val_loss, MS_train_loss, MS_val_loss):
    # plt.style.use("ggplot")
    # plt.figure()
    # plt.subplot(1, 1, 1)
    # plt.title("Accuracy")
    
    # plt.plot(epoch, CL_train_loss, label='OurLager_train', color='b')
    # plt.plot(epoch, CL_val_loss,   label='OurLager_val'  , color='g')
    # plt.plot(epoch, CS_train_loss, label='OurSmall_train', color='m')
    # plt.plot(epoch, CS_val_loss,   label='OurSmall_val'  , color='y')
    # plt.plot(epoch, ML_train_loss, label='V3Lager_train', color='r')
    # plt.plot(epoch, ML_val_loss,   label='V3Lager_val'  , color='c')
    # plt.plot(epoch, MS_train_loss, label='V3Small_train', color='k')
    # plt.plot(epoch, MS_val_loss,   label='V3Small_val'  , color='#F75000')
    
    plt.plot(epoch, CL_train_loss, label='OurLager_train')
    plt.plot(epoch, CL_val_loss,   label='OurLager_val')
    plt.plot(epoch, CS_train_loss, label='OurSmall_train')
    plt.plot(epoch, CS_val_loss,   label='OurSmall_val')
    plt.plot(epoch, ML_train_loss, label='V3Lager_train')
    plt.plot(epoch, ML_val_loss,   label='V3Lager_val')
    plt.plot(epoch, MS_train_loss, label='V3Small_train')
    plt.plot(epoch, MS_val_loss,   label='V3Small_val')
    
    plt.legend(bbox_to_anchor=(1.05, 1.0))
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.tight_layout()
    # plt.savefig(os.path.join(res_dir, 'loss.png'))
    plt.show()
```

```python
def read_value(CL_train_df, CL_val_df,ML_train_df, ML_val_df, CS_train_df, CS_val_df,MS_train_df, MS_val_df):
    epoch = CL_train_df['Step'].values
    CL_train_loss = CL_train_df['Value']
    CL_val_loss = CL_val_df['Value']
    ML_train_loss = ML_train_df['Value']
    ML_val_loss = ML_val_df['Value']
    CS_train_loss = CS_train_df['Value']
    CS_val_loss = CS_val_df['Value']
    MS_train_loss = MS_train_df['Value']
    MS_val_loss = MS_val_df['Value']
    
    return epoch, CL_train_loss, CL_val_loss, ML_train_loss, ML_val_loss, CS_train_loss, CS_val_loss, MS_train_loss, MS_val_loss
```

```python
if __name__ == "__main__":
    model = 'CustomizeLarge'
    dataset = ["cifar10", "cifar100"]
    # root_dir = os.getcwd()
    # file_dir = os.path.join(root_dir, 'file')
    # res_dir = os.path.join(root_dir, 'results')
    for i in dataset:
        CL_train_path = '../實驗/{0}/Loss/CustomizeLarge_SE_Adam_{0}_train.csv'.format(i)
        CL_val_path = '../實驗/{0}/Loss/CustomizeLarge_SE_Adam_{0}_validation.csv'.format(i)
        ML_train_path = '../實驗/{0}/Loss/MobileNetV3Large_SE_Adam_{0}_train.csv'.format(i)
        ML_val_path = '../實驗/{0}/Loss/MobileNetV3Large_SE_Adam_{0}_validation.csv'.format(i)
        CS_train_path = '../實驗/{0}/Loss/CustomizeSmall_SE_Adam_{0}_train.csv'.format(i)
        CS_val_path = '../實驗/{0}/Loss/CustomizeSmall_SE_Adam_{0}_validation.csv'.format(i)
        MS_train_path = '../實驗/{0}/Loss/MobileNetV3Small_SE_Adam_{0}_train.csv'.format(i)
        MS_val_path = '../實驗/{0}/Loss/MobileNetV3Small_SE_Adam_{0}_validation.csv'.format(i)
        
        CL_train_df = pd.read_csv(CL_train_path)
        CL_val_df = pd.read_csv(CL_val_path)
        ML_train_df = pd.read_csv(ML_train_path)
        ML_val_df = pd.read_csv(ML_val_path)
        CS_train_df = pd.read_csv(CS_train_path)
        CS_val_df = pd.read_csv(CS_val_path)
        MS_train_df = pd.read_csv(MS_train_path)
        MS_val_df = pd.read_csv(MS_val_path)
        
        epoch, CL_train_loss, CL_val_loss, ML_train_loss, ML_val_loss, CS_train_loss, CS_val_loss, MS_train_loss, MS_val_loss = \
            read_value(CL_train_df, CL_val_df, ML_train_df, ML_val_df, CS_train_df, CS_val_df, MS_train_df, MS_val_df)
        loss_visualize(epoch, CL_train_loss, CL_val_loss, ML_train_loss, ML_val_loss, CS_train_loss, CS_val_loss, MS_train_loss, MS_val_loss)
```
